# Data Selection & Cleaning - n3

Load preprocessed dataset từ n2, hiển thị info chi tiết để quyết định xóa data nào


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Set pandas options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 50)
pd.set_option('display.max_rows', None)

print("✅ Libraries imported successfully")
```

## 1. Load Preprocessed Data


```python
# Load preprocessed data from n2
preprocessed_path = '../../data/preprocessed/preprocessed_data.csv'
df = pd.read_csv(preprocessed_path)

print(f"✅ Loaded data successfully!")
print(f"   Path: {preprocessed_path}")
print(f"   Rows: {len(df)}")
print(f"   Columns: {len(df.columns)}")
```

## 2. Dataset Overview


```python
print("="*80)
print("DATASET OVERVIEW")
print("="*80)
print(f"\nShape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\nData Types:")
print(df.dtypes)
print(f"\n{'-'*80}")
print("Column Names:")
print(f"{'-'*80}")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2d}. {col}")
```

## 3. Missing Values & Data Quality


```python
print("\n" + "="*80)
print("MISSING VALUES")
print("="*80)

missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing Count': missing.values,
    'Percentage %': missing_pct.values
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if len(missing_df) == 0:
    print("✅ No missing values found!")
else:
    print(missing_df.to_string(index=False))

print(f"\n{'-'*80}")
print("NULL VALUES SUMMARY:")
print(f"{'-'*80}")
print(f"Total cells: {df.size}")
print(f"Null cells: {df.isnull().sum().sum()}")
print(f"Null %: {(df.isnull().sum().sum() / df.size) * 100:.2f}%")
```

## 4. Sample Data 


```python
print("\n" + "="*80)
print("FIRST 20 ROWS - ALL COLUMNS")
print("="*80 + "\n")
print(df.head(20).to_string())
```

## 5. Drop Unnecessary Columns


```python
# Columns to drop (unnecessary for model)
cols_to_drop = [
    'user_name',        # 2. User identifier (not useful for fake news detection)
    'timestamp_post',   # 3. Raw timestamp (already extracted hour/day/month)
    'num_post',         # 10. User statistics
    'num_real',         # 11. User statistics
    'num_fake',         # 12. User statistics
    'post_ratio',       # 13. User statistics (derived)
    'pixel',            # 17. Image properties
    'image',            # 18. Image binary
    'num_image',        # 19. Number of images
    'hour',             # 20. Used to create cyclical features (feat_hour_sin/cos)
    'weekday',          # 21. Used to create cyclical features (feat_day_sin/cos)
    'day',              # 22. Day of month (less useful)
    'month',            # 23. Used to create cyclical features (feat_month_sin/cos)
    'year',             # 24. Year (all same)
    'time',             # 25. Raw time (already encoded)
    'min',              # 26. Minutes (already encoded)
]

print("="*80)
print("DROPPING UNNECESSARY COLUMNS")
print("="*80)
print(f"\nBefore: {df.shape}")
print(f"Columns to drop: {len(cols_to_drop)}")

# Drop columns that exist in dataframe
cols_to_drop_actual = [col for col in cols_to_drop if col in df.columns]
df = df.drop(columns=cols_to_drop_actual)

print(f"\nAfter: {df.shape}")
print(f"✅ Dropped {len(cols_to_drop_actual)} columns")
print(f"\nRemaining columns ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:2d}. {col}")
```

## 6. Save Cleaned Dataset


```python
# Save cleaned dataset
output_path = '../../data/preprocessed/preprocessed_data.csv'
df.to_csv(output_path, index=False)

print("="*80)
print("DATASET SAVED")
print("="*80)
print(f"\n✅ Saved to: {output_path}")
print(f"   Shape: {df.shape}")
print(f"   Rows: {df.shape[0]}")
print(f"   Columns: {df.shape[1]}")
print(f"\n🚀 Dataset cleaned and ready for next phase!")
```
