# S0 - Shuffle Data

**Mục đích**: Trộn dữ liệu và lưu lại để sử dụng cho các bước encoding tiếp theo.

**Output**: `data/encoded/final_data_shuffled.csv`

```python
import pandas as pd
import numpy as np
from pathlib import Path

# Project paths
PROJECT_ROOT = Path.cwd().parent.parent
DATA_DIR = PROJECT_ROOT / 'data'
PREPROCESSED_DIR = DATA_DIR / 'preprocessed'
ENCODED_DIR = DATA_DIR / 'encoded'

# Ensure output directory exists
ENCODED_DIR.mkdir(parents=True, exist_ok=True)

print(f"Project root: {PROJECT_ROOT}")
print(f"Data directory: {DATA_DIR}")
```

```python
# Load data
INPUT_FILE = PREPROCESSED_DIR / 'final_data.csv'
df = pd.read_csv(INPUT_FILE)

print(f"Original data shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nLabel distribution:")
print(df['label'].value_counts())
```

```python
# Shuffle data with fixed random state for reproducibility
RANDOM_STATE = 42

df_shuffled = df.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

print(f"Shuffled data shape: {df_shuffled.shape}")
print(f"\nFirst 5 rows (IDs should be random):")
print(df_shuffled[['id', 'label']].head(10))
```

```python
# Save shuffled data
OUTPUT_FILE = ENCODED_DIR / 'final_data_shuffled.csv'
df_shuffled.to_csv(OUTPUT_FILE, index=False)

print(f"\n✓ Saved shuffled data to: {OUTPUT_FILE}")
print(f"\nFile size: {OUTPUT_FILE.stat().st_size / 1024 / 1024:.2f} MB")
```

```python
# Verify
df_verify = pd.read_csv(OUTPUT_FILE)
print(f"Verification - Shape: {df_verify.shape}")
print(f"Label distribution (should be same as original):")
print(df_verify['label'].value_counts())

# Check for NaN in text columns
print(f"\nNaN counts:")
print(f"  text_tfidf: {df_verify['text_tfidf'].isna().sum()}")
print(f"  text_bert: {df_verify['text_bert'].isna().sum()}")
```
